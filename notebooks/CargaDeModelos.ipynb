# RandomForestClassifier_SF_2

```python

import pickle

# Cargar el modelo
with open("modelos/RandomForestClassifier_SF_2.pkl", "rb") as file:
    RandomForestClassifier_SF_2 = pickle.load(file)

# Mostrar las variables que espera el RandomForestClassifier_SF_2
if hasattr(RandomForestClassifier_SF_2, 'feature_names_in_'):
    print("Variables esperadas por el RandomForestClassifier_SF_2:")
    print(RandomForestClassifier_SF_2.feature_names_in_)
else:
    print("El modelo no tiene el atributo 'feature_names_in_'.")
```

```python
# Ver los parámetros del modelo
print("Parámetros del modelo:")
print(RandomForestClassifier_SF_2.get_params())
```

# best_xgb_08_02

```python

import pickle

# Cargar el modelo
with open("modelos/best_xgb_08_02.pkl", "rb") as file:
    best_xgb_08_02 = pickle.load(file)

# Obtener las características esperadas
if hasattr(best_xgb_08_02, "get_booster"):
    features = best_xgb_08_02.get_booster().feature_names
else:
    features = None

# Mostrar las columnas requeridas
print("Columnas esperadas por el best_xgb_08_02:", features)

# Obtener los parámetros generales del modelo
print("Parámetros generales del modelo XGBoost:")
print(best_xgb_08_02.get_params())
```
